## **Spread**
**Sample Standard Deviation and Variance**

$S = \sqrt{\frac{\sum(X - \bar{X})^2}{n - 1}} \quad \quad \quad S^2 = \frac{\sum(X - \bar{X})^2}{n - 1}$

**Z-Score:**

$Z = \frac{X - \bar{X}}{S}$

<br>

***

<br>

## **One-Sample T-test / Paired T-test**

### **Classic**

**Standard Error:**

$S_{\bar{X}} = \frac{S}{\sqrt{N}}$

**Degrees of Freedom:**

$df = N - 1$

**Test Statistic:**

$T = \frac{\bar{X} - \mu_0}{s_\bar{X}}$

<br>

### **Trimmed**

**Standard Error:**

$s_{\bar{X}_t} = \frac{S_w}{(1-2G) \sqrt{N}} \quad \quad G = \text{Proportion trimmed (on each side)}$

**Degrees of Freedom:**

$df_t = h - 1 \quad \quad \quad h = N - 2 \cdot \lfloor G \cdot N \rfloor$

**Test Statistic:**

$T_t = \frac{(1 - 2G)(\bar{X}_t - \mu_0)}{\frac{S_w}{\sqrt{N}}}$

<br>

***

<br>

## **Two-Sample Independent T-Test**

### **Classic**

**Standard Error (pooled):**

$S_{\bar{X}_2 - \bar{X}_1} = \sqrt{\frac{S^2_{\text{p}}}{n_{\text{2}}} + \frac{S^2_{\text{p}}}{n_{\text{1}}}} \quad \quad S^2_p = \frac{(n_2 - 1) \, S^2_2 \, + \,(n_1 - 1) \, S^2_2}{n_2 \, + \, n_1 \,- 2}$

**Degrees of Freedom:**

$df = n_2 + n_1 - 2$

**Test-Statistic:**

$T = \frac{(\bar{X}_2 - \bar{X}_1) - (\mu_2 - \mu_1)}{S_{\bar{X}_2 - \bar{X}_1}}$

**Effect Size: d**

$d = \frac{\bar{X}_2 - \bar{X}_1}{S_p}$

**Effect Size: g**

$g = d \cdot \left(1 - \frac{3}{4(n_1 + n_2)-9}\right)$

<br>

### **Welch:**

**Standard Error:**

$S_{\bar{X}_2 - \bar{X}_1} = \sqrt{\frac{S^2_2}{n_2} + \frac{S^2_1}{n_1}}$

**Test Statistic:**

$T = \frac{(\bar{X}_2 - \bar{X}_1) - (\mu_2 - \mu_1)}{S_{\bar{X}_2 - \bar{X}_1}}$

**Degrees of Freedom:**

$df = \frac{(q_1 + q_2)^2}{\frac{q^2_1}{n_1-1} + \frac{q^2_2}{n_2-1}} \quad \quad \quad q_j = \frac{S^2_j}{n_j}$

<br>

### **Trimmed Means:**

**Squared-Standard Error (for the jth group):**

$d_j = \frac{(n_j - 1)S^2_{wj}}{h_j(h_j - 1)} \quad \quad \quad h_j = n_j - 2 \cdot \lfloor G \cdot n_j\rfloor$

**Degrees of Freedom:**

$df = \frac{(d_2 + d_1)^2}{\frac{d^2_2}{h_2-1} + \frac{d^2_1}{h_1-1}}$


**Test-Statistic:**

$T_y = \frac{(\bar{X}_{t2} - \bar{X}_{t1}) - (\mu_2 - \mu_1)}{\sqrt{d_2 + d_1}}$

**Confidence Interval:**

$(\bar{X}_{t2} - \bar{X}_{t1}) \pm \sqrt{d_2 + d_1} \cdot T_\text{crit}$

## **Simple Linear Regression**
(i.e., Linear Regression with one predictor)

**Pearson's Correlation Coefficient R:**

$R = \frac{\sum(Z_x \cdot Z_y)}{N - 1}$

**Standard Error of R:**

$S_R = \sqrt{\frac{1 - R^2}{N - 2}}$

**Degrees of Freedom of R:**

$df_R = N - 2$ 

**Slope:**

$b_1 = R \frac{S_y}{S_x}$

**Intercept:**

$b_0 = \bar{Y} - b_1 \cdot \bar{X}$

**Slope and Intercept Standard Error:**

$S_{b_1} = \sqrt{\frac{S^2_{\text{resid}}}{SS_X}} \quad \quad \quad S_{b_0} = \sqrt{\frac{S^2_{\text{resid}} \cdot \sum X^2_i}{N \cdot SS_X}}$

$S^2_{resid} = \frac{\sum (Y - \hat{Y})^2}{N - 2} \quad \quad \quad SS_X = \sum (X_i - \bar{X})^2$

**Degrees of Freedom:**

$df = N - P - 1 \quad \quad \quad P = \text{The number of predictors}$

**Test-Statistics:**

$T_{b_1} = \frac{b_1 - \beta_1}{S_{b_1}} \quad \quad \quad T_{b_0} = \frac{b_0 - \beta_0}{S_{b_0}}$

<br>

## **Multiple Regression and ANOVA**

**Multiple R-Squared:**

$
\begin{align*}
R^2 = \frac{SS_M}{SS_T}
\end{align*}
$

<br>

**Adjusted R-Squared:**

$
\begin{align*}
R^2_\text{adj} = 1 - \frac{(1 - R^2)(N - 1)}{N - P - 1}
\end{align*}
$

<br>

**Total Sum of Squares Formulas ($SS_T$):**

$
\begin{align*}
SS_T &= SS_M + SS_R \\
&= \Sigma (Y_i - \bar{Y}_\text{grand})^2 \\
&= S^2_\text{grand} \cdot (N-1)
\end{align*}
$

$i =$ individual observations 

$df_T = N - 1$

<br>

**Model Sum of Squares Formulas ($SS_M$):**

$\begin{align*}
SS_M &= SS_T - SS_R \\
&= \Sigma (\hat{Y}_i - \bar{Y}_\text{grand})^2 \\
&=\Sigma \ n_j \cdot (\bar{Y}_j - \bar{Y}_\text{grand})^2
\end{align*}$

$j =$ number of levels/groups across predictors

$df_M = j - 1$

<br>

**Main Effect and Interaction Sum of Squares ($SS_A$, $SS_A$, $SS_{A \times B}$):**

$SS_A = \sum\limits_{n=1}^j n_j \cdot (\bar{Y}_j - \bar{Y}_\text{grand})^2$

$SS_B = \sum\limits_{n=1}^j n_j \cdot (\bar{Y}_j - \bar{Y}_\text{grand})^2$

$SS_{A \times B} = SS_M - SS_A - SS_B$

$j =$ levels/groups within the predictor

$df_A = j - 1 \quad \quad df_B = j - 1 \quad \quad df_{A \times B} = df_A \cdot df_B$

<br>

**Residual Sum of Squares Formulas ($SS_R$ or $SS_E$):**

$\begin{align*}
SS_R &= SS_T - SS_M \\
&= \Sigma (Y_i - \hat{Y}_i)^2 \\
&= \Sigma (Y_{ij} - \bar{Y}_j)^2 \\
&= \Sigma S^2_j \cdot (n_j - 1)
\end{align*}$

$df_R = df_T - df_M$

<br>

**Mean-Squares:**

$MS_x = \frac{SS_x}{df_x}$

<br>

**F-stat to Compare Regression Slopes:**

$F = \left( \frac{N - P - 1}{P}\right) \left( \frac{R^2}{1 - R^2}\right) \quad \quad \quad df_1 = P \quad \quad df_2 = N - P - 1$

<br>

**F-stat to Compare Two Hierarchical Regression Models:**

$F_\text{diff} = \left( \frac{N - P_{\text{mod2}} - 1}{P_{\text{diff}}}\right) \left( \frac{R^2_{\text{diff}}}{1 - R^2_{\text{mod2}}}\right) \quad \quad \quad df_1 = P_\text{diff} \quad \quad \quad df_2 = N - P_\text{mod2} - 1$

<br>

**F-stat to test Main Effects and Interactions in an ANOVA**

$F = \frac{MS_x}{MS_R} \quad \quad \quad df_1 = SS_x \quad \quad df_2 = SS_R$

## **Effect Size**

#### **Two Independent Means:**

**Cohen's d:**

$d = \frac{\bar{x}_2 - \bar{x}_1}{S_p}$

$S_p = \sqrt{\frac{(n_1 - 1) \cdot S^2_1 + (n_2 - 1) \cdot S^2_2}{n_1 + n_2 - 1}}$

<br>

**Hedge's g:**

$g = d \cdot \left(1 - \frac{3}{4(n_1 + n_2)-9}\right)$

<br>

#### **Two Dependent Means:**

**Cohen's d:**

$d = \frac{\bar{X}_1 - \bar{X}_2}{S_{rm}}$

$S_{rm} = \frac{S_D}{\sqrt{2(1-R)}}$

$S_D =$ Standard Deviation of the difference scores.

$R =$ Pearson's Correlation Coefficient

<br>

**Hedge's g:**

$g = d \cdot \left(1 - \frac{3}{4(N - 1)-1}\right)$

<br>

#### **ANOVA Main Effects and Interaction**
*Note: A one way ANOVA is just a Two-way ANOVA with a single Main effect and no interaction.*

$\omega^2_x = \frac{SS_x - df_x \cdot MS_R}{SS_T + MS_R}$

$x =$ predictor or interaction

<br>

#### **ANOVA Planned Comparisons in terms of R:**

$R = \frac{T^2}{T^2 + df}$

$df = N - P - 1$

$T =$ The test-statistic for the contrast's $b$ value.

$P =$ the number of predictors (i.e., dummy variables/contrasts)

<br>

#### **ANOVA Pairwise Comparisons in terms of D:**

*see Cohen's d for a Two-Sample Independent T-test.*

## **Chi-Square Test**

$\chi^2 = \sum \frac{(O - E)^2}{E}$

- $O =$ observed frequency

- $E =$ Expected frequency

- Sum across cells.

**Degrees of Freedom for One Variable:**

$df = k-1$

$k =$ Number of categories.

<br>

**Expected Frequencies for Two Variables:**

$E_{ij} = \frac{R_i \cdot C_j}{N}$

- R = row total
- C = Column total

**Degrees of Freedom for Two Variables:**

$df = (R - 1)(C - 1)$

- R = Number of rows in the table.
- C = Number of columns in the table.

## **Miscellaneous**

**MAD-Median Rule:**

$\frac{|X - \text{median}(x)|}{\text{MADN}} > 2.24$

<br>

**Biserial Correlation:**

$R_b = \frac{R_{pb} \cdot \sqrt{p \cdot q}}{u}$

- $R_{pb} =$ Point-Biseral Correlation.
- $p =$ Proportion of cases falling into the largest category.
- $q =$ Proportion of cases falling into the smallest category.
- $u =$ the ordinate (i.e., y-axis value) of a standard normal distribution at $p$.

- Standard Error: $S_{R_b} = \frac{\sqrt{pq}}{u \sqrt{N}}$
- Test Statistic: $Z_{R_b} = \frac{R_b - \rho_b}{S_{R_b}}$

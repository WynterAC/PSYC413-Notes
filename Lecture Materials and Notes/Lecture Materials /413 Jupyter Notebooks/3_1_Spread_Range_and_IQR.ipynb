# Interquartile Range (IQR)

```r
print(quantile(heightData$Father))
```

```r
median(heightData$Father)
```

```r
ggplot(heightData, aes(x = Father)) +
  geom_histogram(colour = "black",
                 fill = "white",
                 binwidth = 1
                 ) +
  theme_bw() +

  #75th Percentile
  annotate("rect",
            xmin = -Inf,
            xmax = 69.6,
            ymin = 0,
            ymax = Inf,
            alpha = 0.3,
            fill = "#F5793A")
```

```r
ggplot(heightData, aes(x = Father)) +
  geom_histogram(colour = "black",
                 fill = "white",
                 binwidth = 1
                 ) +
  theme_bw() +

  #25 Percentile
  annotate("rect",
            xmin = -Inf,
            xmax = 65.8,
            ymin = 0,
            ymax = Inf,
            alpha = 0.3,
            fill = "#A95AA1")
```

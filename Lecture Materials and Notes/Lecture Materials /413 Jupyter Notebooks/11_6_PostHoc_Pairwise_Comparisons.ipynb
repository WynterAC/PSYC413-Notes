p <- c(0.00, 0.0017, 0.0188, 0.0274, 0.1682, 0.3566)

# Bonferroni Correction

$p' = p \cdot k$

```r
p.adjust(p, "bonferroni") # Note: R output adjusts the p-values not the alpha level.
```

# Holm-Bonferroni:

$p' = p \cdot i$

```r
p.adjust(p, "holm")
```

```r
p_ex <- c(0.001, 0.051, 0.07)
```

```r
p_ex * 3:1
```

```r
p.adjust(p_ex, "holm")
```

# Benjamini-Hochberg:

$p' = p \cdot \frac{k}{i}$

```r
p.adjust(p, "BH")
```

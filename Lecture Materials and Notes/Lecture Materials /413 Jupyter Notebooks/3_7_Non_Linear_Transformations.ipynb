```r
install.packages("rcompanion")
```

```r
library(rcompanion)
library(ggplot2)
options(repr.plot.width = 7, repr.plot.height = 5)
```

```r
set.seed(1)
x <- rbeta(5000, 1, 8)
```

```r
ggplot(mapping = aes(x = x)) +
  geom_histogram(fill = "white", colour = "black")
```

```r
ggplot(mapping = aes(sample = x)) +
  stat_qq() +
  stat_qq_line()
```

```r
transformTukey(x, plotit = FALSE, returnLambda = TRUE)
```

```r
t <- x^0.325
```

```r
ggplot(mapping = aes(x = t)) +
  geom_histogram(fill = "pink", colour = "black")
```

```r
ggplot(mapping = aes(sample = t)) +
  stat_qq(colour = "black") +
  stat_qq_line()
```

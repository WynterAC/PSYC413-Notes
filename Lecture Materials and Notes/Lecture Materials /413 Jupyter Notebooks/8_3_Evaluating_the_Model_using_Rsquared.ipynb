```r
library(ggplot2)
options(repr.plot.width = 7, repr.plot.height = 5)
```

```r
heightData <- read.csv("PearsonHeightData.csv")
```

```r
ggplot(heightData, aes(x = Father, y = Son)) +
  geom_point(size = 3) +
  geom_abline(intercept = 33.96, slope = 0.513,
              colour = "cyan4", linewidth = 2)
```

```r
heightData$predictions <- 33.96 + 0.51 * heightData$Father
```

```r
heightData$resid <- heightData$Son - heightData$predictions
```

```r
options(repr.plot.width = 5, repr.plot.height = 5)

ggplot(heightData, aes(x = Father, y = resid)) +
  geom_point(size = 3, shape = 21, colour = "black", fill = "cyan4", stroke = 0.5) +
  geom_hline(yintercept = 0)
```

```r
ggplot(heightData, aes(x = Father, y = resid)) +
  geom_point(size = 3, shape = 21, colour = "black", fill = "cyan4", stroke = 0.5) +
  geom_hline(yintercept = 0) +
  coord_cartesian(xlim = c(64, 71))
```

```r
ggplot(heightData, aes(x = resid)) +
  geom_histogram(colour = "black", fill = "lightcoral") +
  geom_vline(xintercept = 0, colour = "cyan4", linewidth = 2)
```

```r
ggplot(heightData, aes(sample = resid)) +
  stat_qq(shape = 21, colour = "black", fill = "lightcoral", size = 3) +
  stat_qq_line(colour = "cyan4", linewidth = 1)
```

```r
ggplot(heightData, aes(x = Father, y = Son)) +
  geom_point(size = 3) +

  # Linear Model
  geom_abline(intercept = 33.96, slope = 0.513,
              colour = "cyan4", linewidth = 2) +

  #Mean Model
  geom_hline(yintercept = 68.68, colour = "red", linewidth = 2)
```

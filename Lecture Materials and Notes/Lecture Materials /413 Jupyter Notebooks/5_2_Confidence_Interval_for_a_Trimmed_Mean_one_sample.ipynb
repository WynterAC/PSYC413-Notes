# Using R


```r
chipData <- read.csv("chipData.csv")
#chipData
```

```r
# Amount of Trimming 
G <- 0.2
```

```r
#Sample Size
N <- length(chipData$Chips)
N
```

```r
#20% Trimmed Mean
M <- mean(chipData$Chips, tr = G)
M
```

```r
install.packages("WRS2")
library(WRS2)
```

```r
# Winsorized Variance
var <- winvar(chipData$Chips)

# Winsorized Standard Deviation
s <- sqrt(var)
s
```

```r
# Standard Error of the 20% Trimmed Mean
SE <- s / ((1 - 2*G) * sqrt(N))
SE
```

```r
trimse(chipData$Chips)
```

```r
# Degrees of Freedom
h <- N - 2 * floor(G * N)
h
```

```r
df <- h - 1
df
```

```r
# Critical Values
t_crit <- abs(qt(.025, df = df))
t_crit
```

```r
# 95% Confidence Interval
M - t_crit * SE
M + t_crit * SE
```

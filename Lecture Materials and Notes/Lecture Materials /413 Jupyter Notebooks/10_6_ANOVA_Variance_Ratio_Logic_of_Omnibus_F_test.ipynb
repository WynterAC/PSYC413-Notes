```r
clinTrial <- read.csv("ANOVAex_Tidy.csv")
#clinTrial
```

```r
# Grand Mean
grand <- mean(clinTrial$BDIscore)
grand
```

```r
# Total Sum of Squares 
SST <- sum((clinTrial$BDIscore - grand)^2)
SST
```

```r
install.packages("plyr")
```

```r
# Model Sum of Squares
library(plyr)
SSM_data <- ddply(clinTrial, c("Treatment"), summarise,
                  n = length(BDIscore),
                  Mean = mean(BDIscore)
                  )

SSM_data
```

```r
SSM <- sum(SSM_data$n * (SSM_data$Mean - grand)^2)
SSM
```

```r
# Residual Sum of Squares
SSR <- SST - SSM
SSR
```

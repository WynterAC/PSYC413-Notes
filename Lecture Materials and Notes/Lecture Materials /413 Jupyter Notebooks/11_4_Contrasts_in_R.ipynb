```r
clinTrial <- read.csv("ANOVAex_Tidy.csv")
```

```r
# Factor our Predictor Column
clinTrial$Treatment <- factor(clinTrial$Treatment)
```

```r
clinTrial$Treatment
```

```r
levels(clinTrial$Treatment)
```

```r
# Planned Contrasts
cont1 <- c(1, 1, -2)
cont2 <- c(1, -1, 0)
```

```r
cbind(cont1, cont2)
```

```r
contrasts(clinTrial$Treatment) <- cbind(cont1, cont2)
```

```r
# Build the Model
theModel <- lm(BDIscore ~ Treatment, data = clinTrial)
```

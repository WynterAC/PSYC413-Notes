```r
bf <- read.csv("Edible_grains.csv")
#bf
```

# A Simple Model

```r
mod_1 <- lm(rating ~ sugar, data = bf)
mod_1
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)
library(ggplot2)
```

```r
ggplot(bf, aes(x = sugar, y = rating)) +
  geom_point() +
  geom_abline(intercept = mod_1$coefficients[1],
              slope = mod_1$coefficients[2]) +
  xlab("Sugar (g)") + ylab("Rating") +
  ggtitle("Model 1: rating ~ sugar")
```

```r
summary(mod_1)
```

```r
names(bf)
```

# Adding a Second Predictor

```r
mod_2 <- lm(rating ~ sugar + fat, data = bf)
mod_2
```

```r
summary(mod_2)
```

```r
ggplot(bf, aes(x = sugar, y = rating)) +
  geom_point() +
  geom_abline(intercept = mod_2$coefficients[1],
              slope = mod_2$coefficients[2]) +
  xlab("Sugar (g)") + ylab("Rating") +
  ggtitle("Model 2: rating ~ sugar + fat")
```

```r
ggplot(bf, aes(x = fat, y = rating)) +
  geom_point() +
  geom_abline(intercept = mod_2$coefficients[1],
              slope = mod_2$coefficients[3]) +
  geom_smooth(method = "lm", se = FALSE) +
  xlab("Fat (g)") + ylab("Rating") +
  ggtitle("Model 2: rating ~ sugar + fat")
```

```r
ggplot(bf, aes(x = sugar, y = rating)) +
  geom_point(aes(size = fat))
```

```r
ggplot(bf, aes(x = sugar, y = rating)) +
  geom_point(aes(colour = fat), size = 5)
```

```r
ggplot(bf, aes(x = sugar, y = rating)) +
  geom_point(aes(colour = fat, size = fat))
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)
library(ggplot2)
```

```r
heightData <- read.csv("PearsonHeightData.csv")
```

```r
ggplot(heightData, aes(x = Father)) +
  geom_histogram(colour = "black",
                  fill = "white",
                  binwidth = 1)
```

```r
median(heightData$Father)
```

```r
mean(heightData$Father)
```

```r
ggplot(heightData, aes(x = Father)) +
  geom_histogram(colour = "black",
                  fill = "white",
                  binwidth = 1) +
                    
  #median
  geom_vline(xintercept = 67.8,
              linetype = 1,
              linewidth = 1.5,
              colour = "blue") +

  #Mean
  geom_vline(xintercept = mean(heightData$Father),
              linetype = 1,
              linewidth = 1.5,
              colour = "red")
```

```r
heightData <- read.csv("PearsonHeightData.csv")
```

```r
#Model Predictions
heightData$y_hat <- 33.96 + 0.513 * heightData$Father
```

```r
#Residuals
heightData$resid <- heightData$Son - heightData$y_hat
```

```r
#Squared Standard Error of the Residuals
n <- nrow(heightData)
var_resid <- sum(heightData$resid^2) / (n - 2)
var_resid
```

```r
#Squared Standard Deviations of X
SSx <- sum((heightData$Father - mean(heightData$Father))^2)
SSx
```

```r
#Standard Error of the Slope
SE_b1 <- sqrt(var_resid / SSx)
SE_b1
```

```r
#Standard Error of the Intercept
SE_b0 <- sqrt((var_resid * sum(heightData$Father^2)) / (n * SSx))
SE_b0
```

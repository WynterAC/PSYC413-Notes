```r
library(ggplot2)
options(repr.plot.width = 7, repr.plot.height = 5)
```

```r
heightData <- read.csv("PearsonHeightData.csv")
```

```r
ggplot(heightData, aes(x = Father, y = Son)) +
  geom_point(size = 3) +
  geom_abline(intercept = 33.96, slope = 0.513,
              colour = "red", linewidth = 1)
```

```r
ggplot(heightData, aes(x = Father, y = Son)) +
  geom_point(size = 3) +
  geom_smooth(method = "lm", se = FALSE)
```

# R Cheat Codes

```r
lm(Son ~ Father, data = heightData)
```

```r
lm(heightData$Son ~ heightData$Father)
```

```r
myModel <- lm(Son ~ Father, data = heightData)
```

```r
myModel
```

```r
Father <- c(0, 15, 17, 120)
predictions <- data.frame(Father)
predictions
```

```r
print(predict(myModel, newdata = predictions))
```

```r
predictions$Son_pred <- predict(myModel, newdata = predictions)
```

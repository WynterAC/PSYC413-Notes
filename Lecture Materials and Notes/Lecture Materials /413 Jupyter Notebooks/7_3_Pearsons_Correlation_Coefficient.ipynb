# Pearson's Correlation Coefficient

# $$R = \frac{\sum(z_x \cdot z_y)}{N - 1}$$

## Step 1. Standardize the values into z-scores

```r
heightData$Father_z <- (heightData$Father - mean(heightData$Father)) / sd(heightData$Father)
```

```r
heightData$Son_z <- (heightData$Son - mean(heightData$Son)) / sd(heightData$Son)
```

```r
heightData
```

```r
ggplot(heightData, aes(x = Father_z, y = Son_z)) +
  geom_point(size = 3) +
  geom_vline(xintercept = 0, linewidth = 1, colour = "red") +
  geom_hline(yintercept = 0, linewidth = 1, colour = "red")
```

# Step 2. Multiply the Z-scores

```r
heightData$product = heightData$Father_z * heightData$Son_z
heightData
```

# Step 3. Sum the Products

```r
sum(heightData$product)
```

```r
n <- nrow(heightData)
```

```r
R = sum(heightData$product) / (n - 1)
```

```r
R
```

# Pearson's Correlation Coefficient Assumptions

- Both variables must be quantitative not categorical.
- Data needs to have a linear relationship (i.e., not have curved shape).
- There are no outliers.

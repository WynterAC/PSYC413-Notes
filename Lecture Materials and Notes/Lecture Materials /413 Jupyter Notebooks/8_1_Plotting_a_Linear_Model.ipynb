# $$\hat{Y} = b_0 + b_1 X$$

$b_0 =$ y-intercept: the value of y when x = 0.

$b_1 =$ slope: the steepness of the line (rise / run)
- The amount of change in Y for a +1 unit change in X.

$X = $ The predictor.

$\hat{Y} = $ the response / prediction.

# Plotting a Linear Model

```r
b0 <- 0
b1 <- 3
```

```r
x <- c(1,2,3,4,5,6,7,8)
```

```r
y <- b0 + b1 * x
y
```

```r
line_mod <- data.frame(x, y)
line_mod
```

```r
ggplot(line_mod, aes(x = x, y = y)) +
  geom_point(size = 3)
```

```r
ggplot(line_mod, aes(x = x, y = y)) +
  geom_line() +
  geom_point(size = 3)
```

```r
ggplot(line_mod, aes(x = x, y = y)) +
  geom_abline(intercept = b0, slope = b1,
                colour = "purple", linewidth = 3, linetype = "dashed") +
  geom_point(size = 7)
```

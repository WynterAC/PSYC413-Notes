```r
ggplot(mapping = aes(x = x1)) +
  geom_histogram()

ggplot(mapping = aes(x = x2)) +
  geom_histogram()
```

```r
ggplot(mapping = aes(x = D)) +
  geom_histogram(fill = "white", colour = "black") +
  ggtitle("Difference Scores")

ggplot(mapping = aes(sample = D)) +
  stat_qq() +
  stat_qq_line() +
  ggtitle("Difference Scores")
```

```r
heightData <- read.csv("PearsonHeightData.csv")
```

```r
med <- median(heightData$Father)
med
```

```r
devs <- heightData$Father - med
```

```r
abs_devs <- abs(devs)
```

```r
MAD <- median(abs_devs)
MAD
```

```r
MADN <- MAD/0.6745
MADN
```

### Calculating MADN the quick way in R

```r
MADN <- mad(heightData$Father)
MADN
```

# Detecting Outliers

```r
abs(heightData$Father - med) / MADN > 2.24
```

```r
heightData$Father_Outlier <- abs(heightData$Father - med) / MADN > 2.24
#heightData
```

```r
outliers <- subset(heightData, Father_Outlier == TRUE)
outliers
```

```r
bf <- read.csv("Edible_grains_clean.csv")
```

```r
bf <- subset(bf, select = c("name", "rating", "sugar", "fat", "fiber", "producer"))
#bf
```

```r
# Original Model
mod3 <- lm(rating ~ sugar + fat + fiber, data = bf)
mod3sum <- summary(mod3)
```

```r
bf$prod_A <- ifelse(bf$producer == "A", 1, 0)
bf$prod_G <- ifelse(bf$producer == "G", 1, 0)
bf$prod_K <- ifelse(bf$producer == "K", 1, 0)
bf$prod_N <- ifelse(bf$producer == "N", 1, 0)
bf$prod_P <- ifelse(bf$producer == "P", 1, 0)
bf$prod_Q <- ifelse(bf$producer == "Q", 1, 0)
```

```r
bf[1:10, ]
```

```r
mod4 <- update(mod3, ~. + prod_A + prod_G + prod_K + prod_N + prod_P + prod_Q)
```

```r
mod4
```

```r
summary(mod4)
```

```r
mod3sum$r.squared
```

```r
anova(mod3, mod4)
```

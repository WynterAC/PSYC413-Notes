```r
heightData <- read.csv("PearsonHeightData.csv")
```

# Calculating the Standard Deviation

## Step 1. Calculate the Deviations
$x - \mu$

```r
#Mean
M <- mean(heightData$Father)
M

# Number of Observations (i.e., sample size)
N = length(heightData$Father)
N
```

```r
# Deviations
devs <- heightData$Father - M
#devs
```

## Step 2. Square the Deviations
$(x - \mu)^2$

```r
devs_sq <- devs^2
#devs_sq
```

## Step 3. Sum the Squared Deviations
$\sum(x - \mu)^2$

```r
SS <- sum(devs_sq)
SS
```

## Step 4. Divide the sum of squared deviations by the number of observations to obtain the variance
$\sigma^2 = \frac{\sum(x - \mu)^2}{N}$

```r
var <- SS/N
var
```

## Step 5. Take the Square Root of the Variance.
$\sigma = \sqrt{\frac{\sum(x - \mu)^2}{N}}$

```r
s <- sqrt(var)
s
```

```r
sqrt(sum((heightData$Father - mean(heightData$Father))^2) / (N - 1))
```

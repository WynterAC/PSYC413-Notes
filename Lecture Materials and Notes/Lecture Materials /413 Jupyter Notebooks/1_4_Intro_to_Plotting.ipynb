```r
# Create a new column of log 10 values for the x-axis
msleep$brainwt_log10 <- log10(msleep$brainwt)
```

```r
options(repr.plot.width = 10, repr.plot.height = 7)
```

```r
ggplot(data = msleep, aes(x = brainwt_log10, y = sleep_total)) +
  geom_point(size = 4, aes(colour = vore)) +
  geom_smooth(se = FALSE)
```

```r
ggplot(data = msleep, aes(x = brainwt_log10, y = sleep_total)) +
  geom_point(size = 4, aes(colour = vore)) +
  geom_smooth(se = FALSE, method = "lm", linewidth = 3, linetype = 2)
```

```r
ggplot(data = msleep, aes(x = brainwt_log10, y = sleep_total)) +
  geom_point(size = 4, aes(colour = vore)) +
  geom_smooth(se = FALSE, method = "lm", linewidth = 3, aes(colour = vore))
```

```r
ggplot(data = msleep, aes(x = brainwt_log10, y = sleep_total)) +
  geom_point(size = 4, aes(colour = vore)) +
  geom_smooth(se = FALSE, method = "lm", linewidth = 3, aes(colour = vore)) +
  facet_wrap(~ vore) +
  xlab("Log10(Brain Weight (kg))") +
  ylab("Sleep Total (hours)") +
  labs(colour = "Diet")
```

```r
ggplot(data = msleep, aes(x = brainwt_log10, y = sleep_total)) +
  geom_point(size = 4, aes(colour = vore)) +
  geom_smooth(se = FALSE, method = "lm", linewidth = 3, aes(colour = vore)) +
  facet_wrap(~ vore, nrow = 1) +
  xlab("Log10(Brain Weight (kg))") +
  ylab("Sleep Total (hours)") +
  labs(colour = "Diet")
```

```r
ggplot(data = msleep, aes(x = brainwt_log10, y = sleep_total)) +
  geom_point(size = 4, aes(colour = vore)) +
  geom_smooth(se = FALSE, method = "lm", linewidth = 3, aes(colour = vore)) +
  facet_wrap(~ vore, nrow = 1) +
  xlab("Log10(Brain Weight (kg))") +
  ylab("Sleep Total (hours)") +
  labs(colour = "Diet") +
  theme_bw()
```

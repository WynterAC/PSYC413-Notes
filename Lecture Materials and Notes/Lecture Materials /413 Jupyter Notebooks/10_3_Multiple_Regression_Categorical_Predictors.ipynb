```r
bf <- read.csv("Edible_grains_clean.csv")
```

```r
bf = subset(bf, select = c("name", "age", "rating", "sugar", "fat", "fiber"))
#bf
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)
library(ggplot2)
```

```r
ggplot(bf, aes(x = sugar, y = rating)) +
  geom_point(size = 3, aes(colour = age, shape = age)) +
  xlab("Sugar (g)") + ylab("Rating")
```

```r
bf$kids <- ifelse(bf$age == "kids", 1, 0)
#bf
```

```r
mod4 <- lm(rating ~ sugar + fat + fiber + kids, data = bf)
mod4
```

```r
summary(mod4)
```

```r
mod3 <- lm(rating ~ sugar + fat + fiber, data = bf)
```

```r
#summary(mod3)
```

```r
anova(mod3, mod4)
```

```r
ggplot(bf, aes(x = sugar, y = rating)) +
  geom_point(size = 3, aes(shape = age)) +
  xlab("Sugar (g)") + ylab("Rating") +

  # Kids Line
  geom_abline(intercept = 53.03,
              slope = -1.95, 
              colour = "blue") +
                
  #Adults Line
  geom_abline(intercept = 53.2,
              slope = -1.95, 
              colour = "red")
```

```r
ggplot(bf, aes(x = sugar, y = rating)) +
  geom_point(size = 3, aes(shape = age)) +
  xlab("Sugar (g)") + ylab("Rating") +

  # Kids Line
  geom_abline(intercept = 53.03,
              slope = -1.95, 
              colour = "blue") +
                
  #Adults Line
  geom_abline(intercept = 53.2,
              slope = -1.95, 
              colour = "red") +
  
  coord_cartesian(xlim = c(0, 4), ylim = c(48, 52))
```

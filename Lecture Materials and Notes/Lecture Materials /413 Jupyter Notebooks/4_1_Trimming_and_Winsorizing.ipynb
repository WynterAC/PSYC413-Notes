```r
set.seed(1)
x <- rbeta(1000, 8, 1)
```

```r
library(ggplot2)
options(repr.plot.width = 7, repr.plot.height = 5)
```

```r
ggplot(mapping = aes(x = x)) +
  geom_histogram(colour = "black", fill = "white")
```

```r
mean(x)
sd(x)
```

```r
median(x)
```

```r
# Trimmed Mean
mean(x, tr = 0.2)
```

```r
install.packages("WRS2")
library(WRS2)
```

```r
# Winsorized Variance
winvar(x, tr = 0.2)
```

```r
# Winsorized Standard Deviation
sqrt(winvar(x, tr = 0.2))
```

# Standardizing the Father Height Data With Z-Scores

```r
heightData <- read.csv("PearsonHeightData.csv")
```

```r
ggplot(heightData, aes(x = Father)) +
  geom_histogram(colour = "black",
                  fill = "white",
                  bins = 30
                  )
```

```r
M <- mean(heightData$Father)
s <- sd(heightData$Father)
M
s
```

```r
heightData$Father_Z <- (heightData$Father - M) / s
#heightData
```

```r
ggplot(heightData, aes(x = Father_Z)) +
  geom_histogram(colour = "black",
                  fill = "steelblue2",
                  bins = 30
                  )
```

```r
M_Z <- mean(heightData$Father_Z)
M_Z
```

```r
round(M_Z, 3)
```

```r
# What is the probability of being 72 inches tall or less.
pnorm(q = 72, mean = 67.69, sd = 2.75)
```

```r
# What is the probability of being 72 inches tall or greater.
1 - pnorm(q = 72, mean = 67.69, sd = 2.75)

pnorm(q = 72, mean = 67.69, sd = 2.75, lower.tail = FALSE)
```

```r
# What is the probability of a value falling below the mean?
pnorm(q = 67.69, mean = 67.69, sd = 2.75)
```

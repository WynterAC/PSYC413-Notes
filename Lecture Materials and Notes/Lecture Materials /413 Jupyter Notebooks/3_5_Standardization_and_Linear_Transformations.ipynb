```r
library(ggplot2)
options(repr.plot.width = 7, repr.plot.height = 5)
```

```r
heightData <- read.csv("PearsonHeightData.csv")
```

```r
heightData$Father_Z <- (heightData$Father - mean(heightData$Father)) / sd(heightData$Father)
```

```r
#heightData
```

```r
heightData$Father_IQ <- heightData$Father_Z * 15 + 100
```

```r
mean(heightData$Father_IQ)
sd(heightData$Father_IQ)
```

```r
ggplot(heightData, aes(x = Father_IQ)) +
  geom_histogram(fill = "white", colour = "black")
```
